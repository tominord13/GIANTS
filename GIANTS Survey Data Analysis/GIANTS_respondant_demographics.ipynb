```python
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.pyplot as plt
import pandas as pd

def plot_distribution(df, column_name, title, color, ordered_categories=None):
    if ordered_categories:
        # Convert the column to a categorical type with the specified order
        df[column_name] = pd.Categorical(df[column_name], categories=ordered_categories, ordered=True)
    
    # Check the value counts for the column
    distribution = df[column_name].value_counts().sort_index()

    # Plot the distribution
    plt.figure(figsize=(8, 6))
    distribution.plot(kind='bar', color=color)
    plt.title(title)
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Define the logical order for age categories
age_order = ['Young', 'Middle-Aged', 'Old', 'Unknown Aged']

# Define the logical order for family size categories
family_size_order = ['Single Person Household', 'Couple', 'Family (3-5)', 'Big Family (5+)', 'Unknown']

# Plotting for market
plot_distribution(df, 'market', 'Market Distribution', 'green')

# Plotting for city type
plot_distribution(df, 'city_type', 'City Type Distribution', 'blue')

# Plotting for gender
plot_distribution(df, 'gender', 'Gender Distribution', 'purple')

# Plotting for mode (primary transportation mode)
plot_distribution(df, 'mode', 'Primary Mode of Transport Distribution', 'red')

# Plotting for age with specific order
plot_distribution(df, 'age', 'Age Distribution', 'orange', ordered_categories=age_order)

# Plotting for income
plot_distribution(df, 'income', 'Income Distribution', 'yellow')

# Plotting for family size with specific order
plot_distribution(df, 'family_size_category', 'Family Size Distribution', 'cyan', ordered_categories=family_size_order)

# Print unique values for family size category
print(df['family_size_category'].unique())
```
